```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the Excel file
file_path = "OR_AE2_Project_Adjusted.xlsx" #Takes 108 seconds
df = pd.read_excel(file_path, engine='openpyxl')

df.head()
```

```python
# Controlla se la colonna "Number of Attendances" esiste e che sia numerica
if 'Number_Of_Attendances' in df.columns and pd.api.types.is_numeric_dtype(df['Number_Of_Attendances']):
    # Crea il dataset esteso replicando ogni riga in base a 'Number of Attendances'
    print("Still working...")
    df_expanded = df.loc[df.index.repeat(df['Number_Of_Attendances'])].reset_index(drop=True)
    
    # Rimuove la colonna "Number of Attendances" se non necessaria
    df_expanded = df_expanded.drop(columns=['Number_Of_Attendances'])
    
    print(f"\nDimensioni del dataset originale: {df.shape}")
    print(f"Dimensioni del dataset esteso: {df_expanded.shape}")
else:
    print("Errore: La colonna 'Number of Attendances' non esiste o non Ã¨ numerica.")
```
